# 16-splitting a string into multiple column

```python
#import libraries
df= pd.DataFrame({'name':['arslan ali', 'masood khan', 'tayyab gujjar', 'haseeb jhammos'],
              'location':['faislabad, pakistan','gujranwala, pakistan','lahore, pakistan','rahimyarkhan, pakistan']})

df
```

```python
#splitting a column into two column
df.name.str.split(' ', expand=True)
```

```python
#adding those splits into new column
df[["first_name", "last_name"]]=df.name.str.split(' ', expand=True)
df
```

# assignments > expand location

```python
#splitting a column into two column
df.location.str.split(' ', expand=True)
```

```python
df[["city_name", "country_name"]]=df.location.str.split(', ', expand=True)
df
```

```python
#refine data manipulation
df=df[['first_name','last_name', 'city_name','country_name']]
df
```

# 23-reshaping a dataframe

```python
  fasla= pd.DataFrame([['12345', 100, 200,300], ['34567', 400,500,600], ['67890', 700, 800, 900]],
                                        columns = ['zip', 'factory', 'warehouse', 'retail'])
                                        
fasla.head()
```

```python
fasla.head().T
```

```python
fasla_long=fasla.melt(id_vars='zip', var_name='location_type', value_name='distance')
fasla_long
```

```python
fasla_long.dtypes
```

```python
import seaborn as sns
sns.barplot(x='zip', y='distance', hue='location_type', data=fasla_long)
```

```python
sns.barplot(x='zip', y='factory', data=fasla)
```

 # 03 how to rename colums

```python
df= pd.DataFrame({'A colum': [1,3,4,5,6,77,77,54,22,3], 'B colum': [ 3,6,7,8,55,3,22,11,56,7]})
df
```

```python
df.rename(columns={'A colum': 'col_a', 'B colum': 'col_b'}, inplace=True)
df
```

```python
#rename columns
df.columns=['col_aa', 'col_ba']
df
```

```python
# to replace any character or string
df.columns=df.columns.str.replace('_', ' ')
df
```

```python
df.columns=df.columns.str.replace(' ', '_')
df
```

```python
# adding prefix to columns
df = df.add_prefix('col_')
df
```

```python
# adding suffix to columns
df = df.add_suffix('baba_')
df
```

```python
# adding suffix to columns
df = df.add_suffix('baba_')
df
```

```python
df.columns= ['col_a', 'col_b']
df
```

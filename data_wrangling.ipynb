```python
ks4= kashti[["age in days", "fare"]]
ks4.head()
```

## - the above data is in wide range and its hard to compare we need to normalize
## - normalization change the values from zero to one(now both values have the same influence on our model)

# method of normalization
- simple feature scaling
     - x(new)= x(old)/x(max)

- min - max method
- Z-score(standard score) -3 to +3
- log tranformation

```python
# simple feature scaling
ks4['fare']= ks4['fare']/ks4['fare'].max()
ks4['age in days']= ks4['age in days']/ks4['age in days'].max()
ks4.head()
```

```python
# min max method

ks4['fare']= (ks4['fare']-ks4['fare'].min())/(ks4['fare'].max()-ks4['fare'].min())
ks4.head()
```

```python
# z-score method
ks4['fare']= (ks4['fare']-ks4['fare'].mean())/ ks4['fare'].std()
ks4.head()
```

```python
# log transformation method
ks['fare']= np.log(ks['fare'])
ks.head()
```

# binning
- grouping of data into smaller number of values (bins)
- convert numeric into categories(jawan, bachy, boorhy)or 1-16 or 17-30 etc
- to have better understanding of groups
    -  low vs mid vs high price
    

```python
bins= np.linspace(min(kashti['age']), max (kashti['age']), 15000)
age_group= ["bachy", "jawan", "boorhy"]
kashti['age']=pd.cut(kashti['age'], bins, labels=age_group, include_lowest=True)
kashti['age']
# how this will change the name in dataset based on grouping
```
